```python
import pandas as pd
import matplotlib.pyplot as plt
import re

# Specify your .out file path
file_path = 'energy.out'

# Initialize lists to store data
data = {'BOND': [], 'ANGLE': [], 'DIHED': []}

# Read data from the .out file
with open(file_path, 'r') as file:
    for line in file:
        matches = re.findall(r'(\S+)\s*=\s*([\d.]+)', line)
        for match in matches:
            key, value = match
            data[key].append(float(value))

# Create a DataFrame
df = pd.DataFrame(data)
print (df)
# Create a plot
plt.figure(figsize=(10, 6))

# Plot each column with different colors
plt.plot(df['BOND'], label='BOND', color='blue')
plt.plot(df['ANGLE'], label='ANGLE', color='orange')
plt.plot(df['DIHED'], label='DIHED', color='green')

plt.title('Energy Values')
plt.xlabel('Data Point')
plt.ylabel('Value')
plt.legend()

plt.show()
```
